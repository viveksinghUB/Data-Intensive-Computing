To select specified columns. Multiple way of zooming in the columns

```r
head(select(flights, year, month, day)) #zooms into year, month and day
head(select(flights, year:day))#selects all the columns between year and day
head(select(flights, -(year:day)))# select all columns except those lying between year and day
```

There are a number of helper functions you can use within select(), like starts_with(), ends_with(), matches() and contains(). 

```r
head(select(flights, matches("dep") ))
head(select(flights, contains("dep") ))
```

We can use select to slice and rename a column although morew convenient function will be rename

```r
head(select(flights, tail_num = tailnum))#slices the table 
head(rename(flights, tail_num = tailnum))#retains the table
```

To extract distinct rows i.e. to find unique values in a table . (This is very similar to base::unique() but should be much faster.)

```r
head(distinct(flights, tailnum))#picks unique values
head(distinct(flights, origin, dest))#picks unique combination of origin and destination.
```

Add new columns with mutate()<br\>
The key difference between mutate() and transform() is that mutate allows you to refer to columns that youve just created:

```r
head(mutate(flights,
  gain = arr_delay - dep_delay,
  gain_per_hour = gain / (air_time / 60)
))
#here we are calculating anf re using gain again
```

If you only want to keep the new variables, use transmute():`

```r
head(transmute(flights,
  gain = arr_delay - dep_delay,
  gain_per_hour = gain / (air_time / 60)
)) # keeps only the new columns
```

We can do calculation like min(), max(), mean(), sum(), sd(), median(), and IQR() on a column to give one value using summarise.


```r
head(summarise(flights,
  delay = mean(dep_delay, na.rm = TRUE)))
#calculate the  mean of the total delay and retuns a new column with the delay
```

```r
#group the flights by tail number
by_tailnum <- group_by(flights, tailnum)
#calculate the mean of distance and the mean of delay of eacch group
delay <- summarise(by_tailnum,
  count = n(),
  dist = mean(distance, na.rm = TRUE),
  delay = mean(arr_delay, na.rm = TRUE))
#pick only those rows which have the delay > 20 and dist < 2000
delay <- filter(delay, count > 20, dist < 2000)
head(delay)
```

n(): the number of observations in the current group<br\>
n_distinct(x):the number of unique values in x.<br\>
first(x), last(x) and nth(x, n) - these work similarly to x[1], x[length(x)], and x[n] but give you more control over the result if the value is missing.

```r
#find the number of planes and the number of flights that go to each possible destination:
group_by(flights, dest)%>%
summarise(planes = n_distinct(tailnum),#total number of distinct flights to that dest
  flights = n() # total number of flights to the dest
)
```

```r
daily <- group_by(flights, year, month, day)
(per_day   <- summarise(daily, flights = n()))
```

Chaining for better redability <br\>
x %>% f(y) turns into f(x, y) so you can use it to rewrite multiple operations that you can read left-to-right, top-to-bottom:

```r
head(flights %>%
  group_by(year, month, day) %>%
  select(arr_delay, dep_delay) %>%
  summarise(
    arr = mean(arr_delay, na.rm = TRUE),
    dep = mean(dep_delay, na.rm = TRUE)
  ) %>%
  filter(arr > 30 | dep > 30))
```
